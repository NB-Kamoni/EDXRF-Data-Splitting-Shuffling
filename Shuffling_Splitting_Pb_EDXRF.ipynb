# Shuffling samples

Samples with concentrations levels above maximum allowable limits (MAL) were labebeled separate from those below MAL to automate the labeling process. The labeled data sets were then marged after labelling.

Before splitting the samples into the training, testing and validation sets it is crucial to shuffle the samples to prevent bias in placement of categorical variables.

```python
import csv
import random

# Open the CSV file in read mode
with open('Pb_edxrf.csv', 'r') as file:
  # Read the contents of the file into a list of rows
  rows = list(csv.reader(file))

# Shuffle the rows
random.shuffle(rows)

# Open the CSV file in write mode
with open('Pb_edxrf.csv', 'w', newline='') as file:
  # Write the shuffled rows back to the file
  writer = csv.writer(file)
  writer.writerows(rows)
```

#Splitting the shuffled data into training, validation and testing sets

60% of the data was allocated to the training set, testing and training sets each had 20% of the total.

```python
!pip install sklearn
import pandas as pd
from sklearn.model_selection import train_test_split

# Loading csv data
df = pd.read_csv('Pb_edxrf.csv')

print(df.to_string())

# Splitting csv data
train, test = train_test_split(df, test_size=0.2, random_state=0)
train, validation = train_test_split(train, test_size=0.2, random_state=0)

# Saving train and test files
train.to_csv('Pb_edxrf_training.csv', index=False)
test.to_csv('Pb_edxrf_testing.csv', index=False)
validation.to_csv('Pb_edxrf_validating.csv', index=False)
```
